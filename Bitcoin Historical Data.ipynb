```python


plt.figure(figsize=(15,5))
plt.plot(df['Close'], label='Close Price', color='blue')
plt.title('BTC/USD Close Price Over Time')
plt.xlabel('Datetime')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Volume'], label="Volume", color='orange')
plt.title("BTC/USD Volume Over Time")
plt.xlabel("Datetime")
plt.ylabel("Volume")
plt.legend()
plt.grid(True)
plt.show()
```

# with volume 

```python
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()
plt.figure(figsize=(15,5))
plt.plot(df['Close'], label = 'Close', alpha=0.5)
plt.plot(df['SMA_50'], label = 'SMA 50', color = 'green')
plt.plot(df['SMA_200'], label='SMA 200',color = 'red')
plt.title("BTC/USD Close Price with SMA 50 & 200")
plt.xlabel("Datetime")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
columns_to_plot = ['Open', 'High', 'Low', 'Close', 'Volume']

for col in columns_to_plot:
    plt.figure(figsize=(14, 5))
    plt.scatter(df.index, df[col], s=1)
    plt.title(f'{col} Over Time')
    plt.xlabel('Datetime')
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Group by 

```python
grouped = df.groupby('date')[["Open", 'High','Low', 'Close', 'Volume']].mean()


grouped.sort_index(inplace=True)


for col in grouped.columns: 
    plt.figure(figsize=(14,5))
    plt.scatter(grouped.index, grouped[col], s=5)
    plt.title(f'Daily Average {col} Over Time')
    plt.xlabel("Date")
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
grouped = df.groupby('date')[["Open", 'High','Low', 'Close', 'Volume']].sum()


grouped.sort_index(inplace=True)


for col in grouped.columns: 
    plt.figure(figsize=(14,5))
    plt.scatter(grouped.index, grouped[col], s=5)
    plt.title(f'Daily Sum {col} Over Time')
    plt.xlabel("Date")
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
grouped = df.groupby('date')[["Open", 'High','Low', 'Close', 'Volume']].max()


grouped.sort_index(inplace=True)


for col in grouped.columns: 
    plt.figure(figsize=(14,5))
    plt.scatter(grouped.index, grouped[col], s=5)
    plt.title(f'Daily max {col} Over Time')
    plt.xlabel("Date")
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
